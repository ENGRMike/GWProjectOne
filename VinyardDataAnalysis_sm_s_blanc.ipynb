```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s_blanc_df = pd.read_csv("/Users/scott/Bootcamp Lessons and Exercises/Project 1 work/sauvignon_blanc.csv")
```

```python
cols = list(s_blanc_df.columns)
cols.remove('Unnamed: 0')
new_df = s_blanc_df.loc[:,cols]
```

```python
new_df.head()
```

```python
new_df.groupby(["country"])["points", "price"].max()
```

```python
plt.figure(figsize=(18,8))
plt.scatter(x='price',y='points', data = new_df)
plt.title('Pinot Noir Ratings by Price')
plt.grid()
plt.show()
```

```python
#SNS Violin plot, width shows how many wines are at that point value
plt.figure(figsize=(18,8))
plot = sns.violinplot(x='country', y = 'points', data = new_df, size = 100)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.scatter(x='country',y='price', data = new_df)
plt.xticks(rotation="vertical")
plt.title('Pinot Noir Ratings by Price')
plt.grid()
plt.show()
```
