```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine_data = pd.read_csv('../winemag-data-130k-v2.csv')
```

```python
#create a list of wines that we are going to analyze
target_wines = ['Pinot Noir', 'Sangiovese']

pinot_df = wine_data[wine_data['variety'] == target_wines[0]]
sangio_df = wine_data[wine_data['variety'] == target_wines[1]]
```

```python
sns.lmplot(x='price',y='points', data = pinot_df, fit_reg = False)
plt.title('Pinot Noir Ratings by Price')
plt.grid()
plt.show()
```

```python
sns.lmplot(x='price', y = 'points', data = sangio_df, fit_reg=False)
plt.title('Sangiovese Ratings by Price')
plt.grid()
plt.show()
```

```python
#SNS Violin plot shows how many wines a country has at what rating
pinot_plot = sns.violinplot(x='country', y = 'points', data = pinot_df, size = 60)
pinot_plot.set_xticklabels(pinot_plot.get_xticklabels(), rotation = 90)
plt.show()
```

```python
sangio_plot = sns.violinplot(x='country', y = 'points', data = sangio_df, size = 60)
sangio_plot.set_xticklabels(sangio_plot.get_xticklabels(), rotation = 90)
plt.show()
```
