```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('merlot.csv')
df = df.drop(columns=['Unnamed: 0'])
df = df.dropna(axis=0, how='any')
df.head()
```

```python
plt.figure(figsize=(18,8))
plt.scatter(df['price'], df['points'])
plt.title('Merlot Ratings by Price')
plt.xlabel('Price')
plt.ylabel('Points')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(18,8))
plot = sns.violinplot(x='country', y = 'points', data = df)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.title('Merlot Ratings by Country')
plt.grid()
plt.xlabel('Country')
plt.ylabel('Points')
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.scatter(df['country'], df['price'])
plt.xticks(rotation='vertical')
plt.title('Merlot Prices by Country')
plt.xlabel('Country')
plt.ylabel('Price')
plt.grid()
plt.show()
```
