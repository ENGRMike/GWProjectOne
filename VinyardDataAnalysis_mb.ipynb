```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine_data = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
#created a list of wines that we are going to analyze 
target_wines = ['Pinot Noir','Riesling','Sauvignon Blanc','Sangiovese','White Blend','Red Blend','Merlot','Chardonnay']

#create independent dataframes for each wine type
pinot_df = wine_data[wine_data['variety'] == target_wines[0]]
riesling_df = wine_data[wine_data['variety'] == target_wines[1]]
sauv_df = wine_data[wine_data['variety'] == target_wines[2]]
sangio_df = wine_data[wine_data['variety'] == target_wines[3]]
wblend_df = wine_data[wine_data['variety'] == target_wines[4]]
rblend_df = wine_data[wine_data['variety'] == target_wines[5]]
merlot_df = wine_data[wine_data['variety'] == target_wines[6]]
chard_df = wine_data[wine_data['variety'] == target_wines[7]]
```

```python
sns.lmplot(x='price',y='points', data = pinot_df, fit_reg = False)
plt.title('Pinot Noir Ratings by Price')
plt.grid()
plt.show()
```

```python
#SNS Violin plot, width shows how many wines are at that point value
plot = sns.violinplot(x='country', y = 'points', data = pinot_df, size = 60)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```
